### Program

Our whole source code with development history is available on github: [scala-joinapp](https://github.com/lovasoa/spark-joinapp).

Our program runs on [Spark 2.1.0](http://spark.apache.org/releases/spark-release-2-1-0.html),
and uses the DataFrame API.
The resource manager we use is [YARN](https://hadoop.apache.org/docs/r2.8.0/hadoop-yarn/hadoop-yarn-site/YARN.html)
from [Apache Hadoop 2.8.0](https://hadoop.apache.org/docs/r2.8.0/).
We set the number of executors to 14 and the number of cores per executor to 5.

We are joining the tables **LINEITEM** and **ORDERS** on the *orderkey*.
The query run resembles *Q3* from the TPCH benchmark, with a few simplifications, because we focus only on data joining:

      SELECT
        l_orderkey,
        l_extendedprice,
        o_orderdate
      FROM
        orders,
        lineitem
      WHERE
        o_custkey % 5 = 0 -- selectivity: 1/5
        AND l_orderkey = o_orderkey
        AND o_orderdate < '1995-03-15' -- selectivity: 0.48
        AND l_shipdate > '1995-03-15' -- selectivity: 0.54


We create a bloom filter from the data in the pre-filtered **ORDERS** table, and filter **LINEITEM** using the resulting bloom filter.

We use the default implementation of bloom filters that is available in spark as *DataFrame.stat.bloomFilter(col, expectedNumItems, fpp)*. The value for *expectedNumItems* is obtained using a fixed-time approximate job (*RDD.countApprox(timeout=3000)*). The value for *fpp* (false-postivive percentage) is gotten from a command-line argument given to the program.

We see that everything is slower with a smaller error rate (that was to be awaited). But if we look at the details, we see that a particular phase of the stages is much slower with the small error rate: the __**Scheduler Delay**__ (17 ms median time vs. 4s).

The official documentation says the following about the **Scheduler delay**  :

> Scheduler delay includes time to ship the task from the scheduler to
   the executor, and time to send the task result from the executor to the scheduler. If
   scheduler delay is large, consider decreasing the size of tasks or decreasing the size
   of task results.

In our case, the long scheduler delay is obviously due to the second part: *send the task result from the executor to the scheduler*. For an error rate of 1%, we spend, for each task around **20 times more** time sending the bloom filter to the scheduler then we spend creating it (4s median time vs 0.2 s).

#### Bloom filter job time in function of the __size__ of the filter in bits

According to [Pagh, Pagh & Rao (2005)](http://www.it-c.dk/people/pagh/papers/bloom.pdf), we have: 

$$ size = elements \times 1.44 \times \log_2 { \frac 1 \epsilon } $$

where

 * $size$ is the size of the bloom filter in bits
 * $elements$ is the number of elements in the bloom filter
 * $\epsilon$ is the Bloomfilter error rate

Here, we have
a scale factor of $SF = 150$,
a normal size of the small table $N_{ORDERS} = 1\ 500\ 000$,
and a selectivity of $\sigma = \frac 1 5 \times 0.48$

So, we end up with:

$$ elements = \sigma \times SF \times N_{ORDERS} = 21\ 600\ 000$$

````python
elements = 1/5 * 0.48 * 1500000 * 150
elements
````

````python
def bloomfilter_size(error_rate):
    return elements * 1.44 * np.log2(1/error_rate)
````

````python
times['bloomfilter_size'] = bloomfilter_size(times.index)
times[['bloomfilter_size']].round().sample(3)
````

````python
ax = times.plot(x='bloomfilter_size', y='time_bloomfilter', style='o', figsize=(15,5), title="Time to create the filter vs. size of the filter")
ax.set_xlabel("Size of the bloom filter in 100s of millions of bits")
ax.set_ylabel("Time to create the bloom filter in seconds")
````

It looks like the time taken by the bloom filter creation job depends linearly on the **size** of the bloom filter.

````python
err_bitsize_poly = np.polyfit(times.bloomfilter_size, times.time_bloomfilter, 1)
err_bitsize_poly
````

So every new bit in the bloom filter slows down the bloom filter creation process by about $3 \times 10^{-8}$ seconds, and there is a constant time to create a filter of any size of about $1.3$ seconds.

Let's translate that in term of error rate:

````python
def model_bloomtime(error_rate):
    bloomsize = bloomfilter_size(error_rate)
    return np.polyval(err_bitsize_poly, bloomsize)
````

````python
error_range = times.index
modeled_time = pd.DataFrame({
    'error_rate' : error_range,
    'modeled_time' : model_bloomtime(error_range)
}).set_index('error_rate')
modeled_time.plot(legend=True)
times.time_bloomfilter.plot(style='x', legend=True)
````

##### Model precision

````python
# Mean squared error
((modeled_time.modeled_time - times.time_bloomfilter) ** 2).mean()
````

````python
## Histogram of the model's error
ax = (modeled_time.modeled_time - times.time_bloomfilter).plot.hist(
    bins=[-1,-.75,-.5,-.25,0,.25,.5,.75,1],
    title="Model error")
ax.set_xlabel("Error in the predicted time, in s")
````

So it looks like our model fits the results quite well, the time taken by the creation of a bloom filter depends linearly on the size of the filter.

### Composition of the join job

So, from what we see above, we conclude that the join job is composed of three stages (stage 4, stage 5, and stage 6). The other stages (0,1,2,and 3) correspond to other jobs.

##### Stage 4
Stage 4 is the fastest of the three. It reads the data from the pre-filtered **ORDERS** table, that was previously persisted to memory and disk on the executors. [Source](https://github.com/lovasoa/spark-joinapp/blob/experiments/src/main/scala/Q3_Bloom.scala#L17-L21).

It then emits values for the **`Sort Merge Join`**, where Spark internally sorts the data at the output of this stage by chunks and persists it to itermediate files.

##### Stage 5

This is the stage in charge of emitting values for the **LINEITEM** table. It reads items from the `lineitem.parquet` parquet file, drops values that don’t match our SQL `WHERE` predicate, loads the previously created bloom filter using Spark's broadcast mechanism (based on peer-to-peer communication), and filters items based on this bloom filter using an [UDF](https://jaceklaskowski.gitbooks.io/mastering-apache-spark/content/spark-sql-udfs.html). [Source](https://github.com/lovasoa/spark-joinapp/blob/experiments/src/main/scala/Q3_Bloom.scala)

Finally, it emits values for the same **`Sort Merge Join`** as stage 4, and Spark also internally sorts and persists emitted records to itermediate files. The sort algorithm used is [Timsort](https://en.wikipedia.org/wiki/Timsort), that has the interesting property of working in $O(M \times N \times \log(M))$ on arrays containing $M$ runs of $N$ sorted elements.

##### Stage 6

This is the longest of the three stages, as it involves the **shuffle** step itself, where all the data is redistributed around the cluster.

It is composed of a number of tasks defined by `spark.sql.shuffle.partitions`. In our case, we didn’t change the default, so we have **200 partitions** and one task for each, so **200 tasks** in output of our join.

Each task asks data to all the mappers, and the mappers respond with the records they have whose hash corresponds to the requesting reducer task ID. The task then proceeds to load in memory all the records it reads, sort them, and join the records by pairs, each element of the pair coming from a different original table.

The task finally writes its results to the underlying HDFS file system, under the form of a parquet file. As the parquet file format authorizes it, the final `result.parquet` will in fact be a folder, containing 200 individual files corresponding to the 200 partitions of the result. Writing the file will induce network transfers, as the distributed file system is aware of the location of the task, and writes the data locally, but replicates each block of data to several nodes for failure transparency.

### Model of the join phase

Let's first find out how many executors can run in parallel on our cluster.

 * $ executors = 14$. This is the number of executors running that can accomplish tasks. This is defined by the following configuration variable in spark: `spark.executor.instances`. We set it before running our program, and it does’nt change after the program has started.
 * $ nodes = 5 $. This is the number of computers on which we can run executors.
 * $ cores = 2 \times 8 = 16 $. This is the number of cores available on each physical node of the cluster.
 * $ threads = 5 $. This is the maximum number of cores we give to a single executor. This is set by `spark.executor.cores`.
 

And as each executor can run $threads$ tasks in parallel, we have a total parallelism of $taskParallel$ : 

$$ taskParallel = \min \left( executors \times cores, nodes \times  cores \right) $$

````python
import math

executors = 14
nodes = 5
cores = 2*8
threads = 5
````

````python
taskParallel = min( executors * threads, nodes * cores)
taskParallel
````

So, when we compute a stage with an optimal parallelism on a datastructure that contains $partitions$ partitions, it takes the following time:

$$ stageTime(partitions, taskTime) = taskTime \times \left\lceil \frac {partitions} {taskParallel} \right\rceil $$

````python
stageTime = lambda partitions, taskTime: taskTime * math.ceil(partitions / (execParallel * cores))
````

In our case, stage 4 and 5 run in parallel, but are different and do not have the same execution time: $ taskTime_4 \lt taskTime_5 $.

We can approximate the total running time of these two stages by :

$$ stageTime_{\{4,5\}} = \frac {stageTime_4 \times partitions_4 + stageTime_5 \times partitions_5} {taskParallel} $$
